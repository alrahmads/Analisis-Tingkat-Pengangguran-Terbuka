Menampilkan nilai outliers dari setiap variabel

```python
def detect_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = (data < lower_bound) | (data > upper_bound)
    return outliers

outliers = detect_outliers(data['Tingkat Pengangguran Terbuka (Persen)'])
outlier_rows = data[outliers]
print("\nNilai Outlier Tingkat Pengangguran Terbuka (Persen):")
print(outlier_rows[['Kabupaten/Kota', 'Tingkat Pengangguran Terbuka (Persen)']])

outliers = detect_outliers(data['Jumlah Penduduk Miskin (Ribu)'])
outlier_rows = data[outliers]
print("\nNilai Outlier Jumlah Penduduk Miskin (Ribu):")
print(outlier_rows[['Kabupaten/Kota', 'Jumlah Penduduk Miskin (Ribu)']])

outliers = detect_outliers(data['Harapan Lama Sekolah'])
outlier_rows = data[outliers]
print("\nNilai Outlier Harapan Lama Sekolah:")
print(outlier_rows[['Kabupaten/Kota', 'Harapan Lama Sekolah']])

outliers = detect_outliers(data['Angka Harapan Hidup'])
outlier_rows = data[outliers]
print("\nNilai Outlier Angka Harapan Hidup:")
print(outlier_rows[['Kabupaten/Kota', 'Angka Harapan Hidup']])
```

Mengganti nilai outliers dengan nilai batas bawah jika nilai outliers berada di bawah batas bawah dan nilai batas atas jika nilai outliers berada di atas batas atas

```python
data_baru = data.copy()

def replace_outliers(data_baru):
    # Identifikasi outliers
    Q1 = data_baru.quantile(0.25)
    Q3 = data_baru.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers_below = data_baru < lower_bound
    outliers_above = data_baru > upper_bound

    # Buat salinan DataFrame
    data_baru_copy = data_baru.copy()

    # Ganti nilai outliers dengan batas bawah atau batas atas pada salinan
    data_baru_copy[outliers_below] = lower_bound
    data_baru_copy[outliers_above] = upper_bound

    return data_baru_copy

data_baru['Tingkat Pengangguran Terbuka (Persen)'] = replace_outliers(data_baru['Tingkat Pengangguran Terbuka (Persen)'])
data_baru['Jumlah Penduduk Miskin (Ribu)'] = replace_outliers(data_baru['Jumlah Penduduk Miskin (Ribu)'])
```
